```python
# 1. Normalize counts per cell (CPM-like, to 1e4 total counts per cell by default)
sc.pp.normalize_total(adata, target_sum=1e4)

# 2. Log1p transform (natural log, log(x+1))
sc.pp.log1p(adata)
```
